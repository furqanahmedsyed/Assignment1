3. What is the most significant distinction between a dictionary and a list?
Answer:- Structure:

List: Ordered collection of elements. Elements are accessed by their position or index in the list (which starts at 0).
Dictionary: Unordered collection of key-value pairs. Elements are accessed by their keys.
Accessing Elements:

List: Elements are accessed by their index, like my_list[0] or my_list[1].
Dictionary: Elements are accessed by their keys, like my_dict['key'].
Mutability:

List: Mutable, meaning you can change, add, or remove elements after the list is created.
Dictionary: Mutable, you can modify, add, or remove key-value pairs after the dictionary is created.
Duplication of Keys:

List: No concept of keys; elements are accessed by position, and there can be duplicate elements.
Dictionary: Keys must be unique within a dictionary; attempting to assign a value to an existing key will overwrite the existing value.

7. What is a shortcut for the following code?
if 'color' not in spam:
spam['color'] = 'black'
Answer:-
You can use the setdefault method to achieve the same result in a more concise way. The setdefault method adds a key to the dictionary with a default value if the key is not already present. [code:- spam.setdefault('color', 'black')]
This line of code checks if the key 'color' is already in the spam dictionary. If it is, it does nothing. If the key is not present, it adds the key 'color' with the default value 'black'. This is equivalent to the original if statement and assignment.


8. How do you 'pretty print' dictionary values using which module and function?
Answer:- To pretty print dictionary values in Python, you can use the pprint module (pretty-print). The pprint module provides a pprint function that can be used to print data structures in a more human-readable format.
(code:- from pprint import pprint
spam = {'a': 1, 'b': 2, 'c': {'x': 10, 'y': 20}}
pprint(spam)  )
In this example, the pprint function from the pprint module is used to print the spam dictionary in a well-formatted and readable manner. The output will be indented and structured, making it easier to understand complex nested dictionaries.
If you want to print the dictionary as a string (for logging or other purposes), you can use the pformat function from the same module:
[code:- from pprint import pformat
pretty_string = pformat(spam)
print(pretty_string)    ]
Both pprint and pformat are part of the pprint module, and you can choose the one that fits your needs.
